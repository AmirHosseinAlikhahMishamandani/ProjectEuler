## Problem Statement

<p>The sequence of triangle numbers is generated by adding the natural numbers. So the $7$<sup>th</sup> triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:
$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, \dots$$</p>

Let us list the factors of the first seven triangle numbers:


\begin{align*}
\mathbf{1} & : 1 \\
\mathbf{3} & : 1, 3 \\
\mathbf{6} & : 1, 2, 3, 6 \\
\mathbf{10} & : 1, 2, 5, 10 \\
\mathbf{15} & : 1, 3, 5, 15 \\
\mathbf{21} & : 1, 3, 7, 21 \\
\mathbf{28} & : 1, 2, 4, 7, 14, 28 \\
\end{align*}


We can see that 28 is the first triangle number to have over five divisors.

**Question:** What is the value of the first triangle number to have over five hundred divisors?


# Solution

To solve this problem, we need to write an algorithm that generates triangle numbers and then finds their divisors. The challenge is to find the first triangle number that has over 500 divisors. Let's break down the steps:

## **1. Generate Triangle Numbers**:

A triangle number is formed by adding the natural numbers. The n-th triangle number is given by the sum of the first n natural numbers, which can be calculated using the formula
$n(n+1)\over 2$.

## **2. Find Divisors of a Number**:

To find the divisors of a number, we can iterate from 1 to the square root of the number, checking if the number is divisible by the iterator. If it is, we count both the divisor and the quotient as divisors of the number.

## **3. Counting Divisors**:

To find the first triangle number with over 500 divisors, we keep generating triangle numbers and counting their divisors until we find one with more than 500 divisors.

```python
def count_divisors(n):
    count = 0
    i = 1
    while i * i <= n:
        if n % i == 0:
            count += 2 if i * i != n else 1
        i += 1
    return count

def find_triangle_number(divisor_limit):
    n = 1
    while True:
        triangle_number = n * (n + 1) // 2
        if count_divisors(triangle_number) > divisor_limit:
            return triangle_number
        n += 1

# Find the first triangle number with over 500 divisors
print(find_triangle_number(500))
```
